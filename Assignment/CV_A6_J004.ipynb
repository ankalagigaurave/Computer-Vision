##Perform image edge detection using sobel and canny filtering operations


```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from scipy import ndimage
```

```python
kyoto_1 = cv2.imread('/content/Kyoto.jpg') 
kyoto_6_1 = mpimg.imread('/content/Kyoto.jpg')
kyoto_2 = cv2.imread('/content/Kyoto.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
edges = cv2.Canny(kyoto_1,60,120)
plt.imshow(kyoto_1,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
```

```python
plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image canny'), plt.xticks([]), plt.yticks([])
```

```python
plt.imshow(kyoto_1,cmap = 'gray')
plt.title('Original Image')
```

```python
sobelx = cv2.Sobel(kyoto_1,cv2.CV_64F,1,0,ksize=1) 
sobely = cv2.Sobel(kyoto_1,cv2.CV_64F,0,1,ksize=1) 
```

```python
plt.imshow(sobelx,cmap = 'gray')
plt.title('Edge Image - sobel x')
```

```python
plt.imshow(sobely,cmap = 'gray')
plt.title('Edge Image - sobel x')
```
