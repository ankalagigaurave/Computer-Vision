## Perform morphological gradient, top hat and black hat morphological operations in an image

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from scipy import ndimage
```

```python
arnie_1 = cv2.imread('/content/arnold.jpg') 
arnie_3_1 = mpimg.imread('/content/arnold.jpg')
arnie_2 = cv2.imread('/content/arnold.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(arnie_1,cmap = 'gray')
plt.title('Original Image')
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
gradient = cv2.morphologyEx(arnie_1, cv2.MORPH_GRADIENT, kernel)
plt.imshow(gradient,cmap = 'gray')
plt.title('Gradient filtered Image')
```

```python
kernel = np.ones((21,21),np.uint8)
```

```python
tophat = cv2.morphologyEx(arnie_1, cv2.MORPH_TOPHAT, kernel)
plt.imshow(tophat,cmap = 'gray')
plt.title('Top Hat filtered Image')
```

```python
blackhat = cv2.morphologyEx(arnie_1, cv2.MORPH_BLACKHAT, kernel)
plt.imshow(tophat,cmap = 'gray')
plt.title('Black Hat filtered Image')
```
