## Perform erosion and dilation morphological operations in an image

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from scipy import ndimage
```

```python
tsu_1 = cv2.imread('/content/The Warrior.jpg') 
tsu_3_1 = mpimg.imread('/content/The Warrior.jpg')
tsu_2 = cv2.imread('/content/The Warrior.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(tsu_3_1)
```

```python
plt.imshow(tsu_1,cmap = 'gray')
plt.title('Original Image')
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
erosion = cv2.erode(tsu_1,kernel,iterations = 4)
plt.imshow(erosion)
plt.title('Erosion Morphological operation')
```

```python
dilation = cv2.dilate(tsu_1,kernel,iterations = 4)
plt.imshow(dilation)
plt.title('Dilation Morphological operation')
```
