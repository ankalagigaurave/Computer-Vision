##To perform geometric transformation operations such as image rotation, scaling and translation

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from scipy import ndimage
```

###Scaling with Linear Interpolation

```python
keanu_scaled = cv2.resize(keanu_1,None,fx=1.2, fy=1.2, interpolation = cv2.INTER_LINEAR)
print('Original shape',keanu_1.shape)
print('Scaled Shape',keanu_scaled.shape)
plt.imshow(keanu_scaled)
plt.title('Scaling with Linear Interpolation')
```

###Scaling with Cubic Interpolation

```python
keanu_scaled = cv2.resize(keanu_1,None,fx=1.2, fy=1.2, interpolation = cv2.INTER_CUBIC)
print('Original shape',keanu_1.shape)
print('Scaled Shape',keanu_scaled.shape)
plt.imshow(keanu_scaled)
plt.title('Scaling with Cubic Interpolation')
```

###Scaling with Skewed Size

```python
keanu_scaled = cv2.resize(keanu_1,(450, 400), interpolation = cv2.INTER_CUBIC)
print('Original shape',keanu_1.shape)
print('Scaled Shape',keanu_scaled.shape)
plt.imshow(keanu_scaled)
plt.title('Scaling with Skewed Size')
```

###Rotation

```python
num_rows, num_cols = keanu_1.shape[:2]
rotate_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 90, 1)
keanu_rotate = cv2.warpAffine(keanu_1, rotate_matrix, (num_cols, num_rows))
plt.imshow(keanu_rotate)
plt.title('Rotation with angle : 90 degrees')
```

###Translation

```python
translation_matrix = np.float32([ [1,0,120], [0,1,225] ])
keanu_translate = cv2.warpAffine(keanu_1, translation_matrix, (num_cols, num_rows))
plt.imshow(keanu_translate)
plt.title('Translation with (+120, +225)')
```

###Rotation and Translation

```python
keanu_rotate = cv2.warpAffine(keanu_translate, rotate_matrix, (2*num_cols, 2*num_rows)) 
plt.imshow(keanu_rotate)
plt.title('Rotation and Translation')
```
