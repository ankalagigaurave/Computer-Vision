##Perform opening and closing morphological operations in an image

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from scipy import ndimage
```

```python
tsu_1 = cv2.imread('/content/tsushima.png') 
tsu_3_1 = mpimg.imread('/content/tsushima.png')
tsu_2 = cv2.imread('/content/tsushima.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(tsu_3_1,cmap = 'gray')
plt.title('Original Image')
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
opening = cv2.morphologyEx(tsu_3_1, cv2.MORPH_OPEN, kernel)
plt.imshow(opening)
plt.title('Opening Morphological operation')
```

```python
closing = cv2.morphologyEx(tsu_3_1, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing)
plt.title('Closing Morphological operation')
```
