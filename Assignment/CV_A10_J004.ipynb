##Perform image segmentation in an image

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from scipy import ndimage
```

```python
arnie_1 = cv2.imread('/content/arnold.jpg') 
arnie_3_1 = mpimg.imread('/content/arnold.jpg')
arnie_2 = cv2.imread('/content/arnold.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(arnie_1,cmap = 'gray')
plt.title('Original Image')
```

```python
pic = arnie_1/255
pic_n = pic.reshape(pic.shape[0]*pic.shape[1], pic.shape[2])
pic_n.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0).fit(pic_n)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
```

```python
cluster_pic = pic2show.reshape(pic.shape[0], pic.shape[1], pic.shape[2])
plt.imshow(cluster_pic)
```
