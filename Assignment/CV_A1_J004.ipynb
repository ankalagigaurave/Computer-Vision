##To perform thresholding operation in an image and evaluate its histogram using histogram equalisation

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from scipy import ndimage
```

```python
arnie_1 = cv2.imread('/content/image.jfif', cv2.IMREAD_COLOR) 
arnie_2 = cv2.imread('/content/image.jfif', 0) 
arnie_3 = cv2.imread('/content/image.jfif', cv2.IMREAD_GRAYSCALE) 
arnie1 = mpimg.imread('/content/image.jfif')
```

```python
plot1 = plt.imshow(arnie1)
```

```python
plt.imshow(arnie_1)
plt.title('OpenCV colour scheme')
```

```python
plt.imshow(arnie_3)
plt.title('Grayscale image')
```

```python
hist = cv2.calcHist([arnie_3],[0],None,[256],[0,256])
plt.hist(arnie_3.ravel(),256,[0,256])
plt.title('Histogram for grayscale image')
plt.show()
```

```python
he1 = cv2.equalizeHist(arnie_3)
plt.imshow(he1)
```

```python
res1 = np.hstack((arnie_3, he1)) 
plt.imshow(res1)
```

```python
arnie = cv2.cvtColor(arnie_1, cv2.COLOR_BGR2GRAY) 
```

```python
ret, thresh1 = cv2.threshold(arnie, 120, 255, cv2.THRESH_BINARY) 
ret, thresh2 = cv2.threshold(arnie, 120, 255, cv2.THRESH_BINARY_INV) 
ret, thresh3 = cv2.threshold(arnie, 120, 255, cv2.THRESH_TRUNC) 
ret, thresh4 = cv2.threshold(arnie, 120, 255, cv2.THRESH_TOZERO) 
ret, thresh5 = cv2.threshold(arnie, 120, 255, cv2.THRESH_TOZERO_INV) 
```

```python
plt.imshow(thresh1)
plt.title('Binary Thresholding')
```

```python
hist_t1 = cv2.calcHist([thresh1],[0],None,[256],[0,256])
plt.hist(thresh1.ravel(),256,[0,256])
plt.title('Histogram for Binary Thresholding image')
plt.show()
```

```python
plt.imshow(thresh2)
plt.title('Binary Thresholding Inverted')
```

```python
hist_t2 = cv2.calcHist([thresh2],[0],None,[256],[0,256])
plt.hist(thresh2.ravel(),256,[0,256])
plt.title('Histogram for Binary Thresholding Inverted image')
plt.show()
```

```python
plt.imshow(thresh3)
plt.title('Truncated Thresholding')
```

```python
hist_t3 = cv2.calcHist([thresh3],[0],None,[256],[0,256])
plt.hist(thresh3.ravel(),256,[0,256])
plt.title('Histogram for Truncated Thresholding image')
plt.show()
```

```python
hist_t4 = cv2.calcHist([thresh4],[0],None,[256],[0,256])
plt.hist(thresh4.ravel(),256,[0,256])
plt.title('Histogram for set to 0 Thresholding image')
plt.show()
```
